### Pruebas con imagenes diferentes a las del dataset con la Red Neuronal


```python
import cv2
from ultralytics import YOLO
from google.colab.patches import cv2_imshow

if __name__ == '__main__':
    img_name = "/content/drive/MyDrive/Prueba_img/img18.jpg"
    img = cv2.imread(img_name)
    model = YOLO('/content/drive/MyDrive/data/train_Yolov8n/weights/best.pt')

    # Realizar la predicción
    preds = model.predict(img, imgsz=640)

    # Dibujar las predicciones en la imagen
    img_with_boxes = preds[0].plot()

    # Mostrar la imagen con las detecciones en una ventana
    cv2_imshow(img_with_boxes)

    # Guardar la imagen con las detecciones
    cv2.imwrite(f"{img_name}_x.jpg", img_with_boxes)

```
